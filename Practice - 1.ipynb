(d)Using box plot, determine anomalies/outliers in any two TS variables.

```python
# Plotting Box Plot for Oil_temperature
plt.figure(figsize=(12, 6))
plt.boxplot(df1['Oil_temperature'])
plt.title('Distribution of Oil_temperature')
plt.xlabel('x')
plt.ylabel('Oil_temperature')
plt.show()
```

```python
# Finding out the outliers

# Calculate the IQR and bounds
Q1 = df1['Oil_temperature'].quantile(0.25)
Q3 = df1['Oil_temperature'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = np.logical_or(df1['Oil_temperature'] < lower_bound, df1['Oil_temperature'] > upper_bound)

# Print the outliers
if outliers.any():
    print("The outliers of Oil temperature variable are:")
    print(df1[outliers]['Oil_temperature'])
else:
    print("No outliers found.")


```

```python
# Box plot for H1 variable
plt.figure(figsize=(12, 6))
plt.boxplot(df1['H1'])
plt.title('Distribution of H1')
plt.xlabel('x')
plt.ylabel('H1')
plt.show()
```

```python
# Calculating the Outliers
# Calculate the IQR and bounds
Q1 = df1['H1'].quantile(0.25)
Q3 = df1['H1'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = np.logical_or(df1['H1'] < lower_bound, df1['H1'] > upper_bound)

# Print the outliers
if outliers.any():
    print("The outliers of H1 TS variable are:")
    print(df1[outliers]['H1'])
else:
    print("No outliers found.")

```
